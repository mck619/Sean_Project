make a table in html, using table, tr, th, td, of the chargers roster with columns for name, number, position and draft

```python
import pandas as pd
```

```python
raw = pd.read_html("http://www.ourlads.com/nfldepthcharts/alpharoster/SD")
```

```python
roster = raw[2]
roster.columns
```

```python
roster
```

```python
roster = roster.drop(1, axis=0)
```

```python
roster.iloc[0]
```

```python
roster.columns = roster.iloc[0]
```

```python
roster = roster.drop(0, axis = 0)
```

```python
roster = roster.set_index('#')
```

```python
roster = roster.drop(['DOB','Age','HT','WT', 'School', 'Orig. Team', 'NFL Exp.'], axis = 1)
```

```python
roster.index = pd.to_numeric(roster.index, errors = 'coerce') 
```

```python
roster = roster.drop([34,53,56,59], axis = 0)
```

```python
roster = roster.drop('...')
```

```python
roster = roster.dropna()
```

```python
def undfa(drafted):
    if "CFA" in drafted:
        return "Undrafted"
    else:
        return drafted

roster['Draft Status'] = roster['Draft Status'].apply(undfa)
    
```

```python
roster
```

```python
print(roster.to_html())
```
