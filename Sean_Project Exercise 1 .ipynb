make a table in html, using table, tr, th, td, of the chargers roster with columns for name, number, position and draft

```python
import pandas as pd
```

```python
roster = pd.read_html("http://www.pro-football-reference.com/teams/sdg/2015_roster.htm")
roster[1]
```

```python
roster = roster[1]
roster.columns
```

```python
roster = roster.rename(columns={'No.':'Number', 'Unnamed: 1':'Name', 'Pos':'Position'})

roster = roster.set_index("Number") #sets index to "No." column
```

```python
roster = roster.drop(roster.columns[[1,3,4,5,6,7,8,9,10,12]], axis=1)
```

```python
roster = roster.fillna('Undrafted')
```

```python
print(roster.to_html())
```
