make a table in html, using table, tr, th, td, of the chargers roster with columns for name, number, position and draft

```python
import pandas as pd
```

```python
raw = pd.read_html("http://www.pro-football-reference.com/teams/sdg/2015_roster.htm")
```

```python
roster = raw[1]
roster.columns
```

```python
roster = roster.rename(columns={'No.':'Number', 'Unnamed: 1':'Name', 'Pos':'Position',
                                'Drafted (tm/rnd/yr)': 'Drafted'})

roster = roster.set_index("Number") #sets index to "Number." column
```

```python
roster = roster.drop(roster.columns[[1,3,4,5,6,7,8,9,10,12]], axis=1)
```

```python
roster[] = roster.fillna('UNDRAFTED')
roster['Position'].apply(lambda x: x.upper())
```

```python
roster
```

```python
print(roster.to_html())
```
